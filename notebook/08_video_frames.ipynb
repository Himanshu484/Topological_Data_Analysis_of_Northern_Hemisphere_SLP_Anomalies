```python
import cv2
import os
from PIL import Image
import numpy as np
```

```python
# Get the path of the notebook file
notebook_path = os.path.abspath('__file__')

# Directory containing the notebook file
notebook_directory = os.path.dirname(notebook_path)
```

```python
def create_side_by_side_video(folder1_path, folder2_path, output_video_path, fps=30):
    """
    Create a video by placing images from two folders side by side.
    
    Args:
        folder1_path: Path to first folder containing images
        folder2_path: Path to second folder containing images  
        output_video_path: Path for output video file
        fps: Frames per second for output video
    """
    
    # Get sorted list of image files from both folders
    images1 = sorted([f for f in os.listdir(folder1_path) 
                     if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    images2 = sorted([f for f in os.listdir(folder2_path) 
                     if f.lower().endswith(('.png', '.jpg', '.jpeg'))])
    
    if len(images1) != len(images2):
        print(f"Warning: Folder 1 has {len(images1)} images, Folder 2 has {len(images2)} images")
    
    # Use minimum number of images
    num_frames = min(len(images1), len(images2))
    
    if num_frames == 0:
        print("No images found in one or both folders!")
        return
    
    # Read first pair to determine dimensions
    img1 = cv2.imread(os.path.join(folder1_path, images1[0]))
    img2 = cv2.imread(os.path.join(folder2_path, images2[0]))
    
    # Get dimensions and resize to same height
    h1, w1 = img1.shape[:2]
    h2, w2 = img2.shape[:2]
    
    # Use smaller height to maintain aspect ratio
    target_height = min(h1, h2)
    new_w1 = int(w1 * target_height / h1)
    new_w2 = int(w2 * target_height / h2)
    
    # Combined frame dimensions
    frame_width = new_w1 + new_w2
    frame_height = target_height
    
    # Initialize video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))
    
    print(f"Creating video with {num_frames} frames...")
    
    for i in range(num_frames):
        # Read and resize images
        img1 = cv2.imread(os.path.join(folder1_path, images1[i]))
        img2 = cv2.imread(os.path.join(folder2_path, images2[i]))
        
        img1_resized = cv2.resize(img1, (new_w1, target_height))
        img2_resized = cv2.resize(img2, (new_w2, target_height))
        
        # Concatenate horizontally
        combined_frame = np.hstack((img1_resized, img2_resized))
        
        # Write frame to video
        out.write(combined_frame)
        
        if (i + 1) % 10 == 0:
            print(f"Processed {i + 1}/{num_frames} frames")
    
    # Release everything
    out.release()
    cv2.destroyAllWindows()
    
    print(f"Video saved as: {output_video_path}")
```

```python
# Usage example:
create_side_by_side_video(
    folder2_path=notebook_directory+"/Data/feature_traj_1_sub",
    folder1_path=notebook_directory+"/Data/pd_traj_1_sub",
    output_video_path=notebook_directory+"/Data/traj_1_sub.mp4",
    fps=10
)
```
